```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


year = ['2013-14', '2014-15', '2015-16', '2016-17', '2017-18', '2018-19','2019-20']

data = pd.DataFrame()
for i in year:
    dt = pd.read_excel(r"D:\MSBA\课本\BUDT704\TeamAssignment\UMD Alumni Association Dataset.xlsx", sheet_name = i)
    print(dt.shape)
    data = data.append(dt, ignore_index = True)
    
data.head()
```

```python
#Split the characters,find the type of activity and participants by Group Description
ActType1 = data['Group Description'].str.split("- ",1).str.get(0)
ActType2 = ActType1.str.split("-",1).str.get(0)
data["ActType"] = ActType2.str.split(" ",1).str.get(1)
data["Participants"] = data['Group Description'].str.split("-",1).str.get(1)
data.head()
```

```python
#Build a new table,delete useless information
df = data[["Event Name","ActType","Participants","Average Age","Participated","First Time Attendees","Percentage First Time Attendees","Major Prospects","Percentage Major Prospect"]]
df
```
